```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv('wta_matches_2016.csv')
df = df.loc[:, ['tourney_name', 'winner_name', 'winner_ioc', 'loser_name', 'loser_ioc']]
df.head()
```

```python
data = list(df['winner_ioc'])
data[:3]
```

```python
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df.astype(np.int8).head()
```

```python
df1 = apriori(df, min_support=0.1, use_colnames=True)

rules = association_rules(df1, metric="lift", min_threshold=1)
rules
```
